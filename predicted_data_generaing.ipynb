```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
import pandas as pd
```

```python
data = pd.read_csv('sales_data_sample.csv', encoding='latin1')
```

```python
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'], format='%m/%d/%Y %H:%M')
```

```python
monthly_sales = data.groupby(pd.Grouper(key='ORDERDATE', freq='M'))['SALES'].sum()
```

```python
model = ExponentialSmoothing(
    monthly_sales,
    trend='add',
    seasonal='add',
    seasonal_periods=12
).fit()
```

```python
forecast = model.forecast(12)
```

```python
future_dates = pd.date_range(
    start=monthly_sales.index[-1] + pd.DateOffset(months=1),
    periods=12,
    freq='M'
)
```

```python
forecast_df = pd.DataFrame({
    'ORDERDATE': future_dates,
    'SALES': forecast.values
})
```

```python
forecast_df.to_csv('monthly_sales_forecast2.csv', index=False)
```
